```python
data = pd.read_excel('Files/transactions_split.xlsx', sheet_name=None)
```

```python
data.keys()
```

```python
txs = data['txs']
txs_total = data['txs_total']
```

```python
txs = txs.merge(txs_total)
```

```python
customers = data['customers']
```

```python
customers
```

```python
txs.head(1)
```

```python
cust_totals = txs.groupby('customer_id', as_index = False)\
                                                .agg(Total_orders = ('transaction_id', 'count'),
                                                avg_amnt = ('amount', 'mean'),
                                                first_purchase = ('date', 'min')).round(1)
```

```python
customers.merge(cust_totals, right_on = 'customer_id', left_on = 'id')
```
