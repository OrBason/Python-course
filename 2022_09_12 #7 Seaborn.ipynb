## Resample

```python
orders = pd.read_csv('Files/Global Orders.csv')
orders.set_index('Order Date').plot(y = 'Profit', figsize = (15,4));
```

```python
orders = pd.read_csv('Files/Global Orders.csv')
orders.set_index('Order Date').sort_index().plot(y = 'Profit', figsize = (15,4));
```

```python
orders = pd.read_csv('Files/Global Orders.csv')
orders['Order Date'] = pd.to_datetime(orders['Order Date'], format = '%d/%m/%Y')
orders.set_index('Order Date').sort_index().plot(y = 'Profit', figsize = (15,4));
```

```python
orders.groupby(orders['Order Date'].dt.year)['Profit'].sum().plot()
```

```python
orders.set_index('Order Date', inplace = True)
```

```python
orders.sort_index(inplace = True)
```

```python
orders.resample('W')[['Profit', 'Sales', 'Shipping Cost']].sum()
```

```python
orders['Profit'].resample('D').sum().plot(figsize = (15,4));
orders['Profit'].resample('W').sum().plot(figsize = (15,4));
orders['Profit'].resample('M').sum().plot(figsize = (15,4));
```
