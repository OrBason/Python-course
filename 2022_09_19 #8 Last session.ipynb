```python
flights = pd.read_csv('Flights-Data.csv')
flights['flight_date'] = pd.to_datetime(flights['flight_date'], format = '%Y%m%d')
```

```python
flights.info()
```

```python
flights[['cancelled', 'diverted']].drop_duplicates()
```

```python
flights['cancelled'].value_counts()
```

```python
flights['cancelled'] = flights['cancelled'].apply(lambda v: False if v in ['F', '0', 'FALSE'] else True)
flights['diverted'] = flights['diverted'].apply(lambda v: False if v in ['F', '0', 'FALSE'] else True)
```

```python
flights.info()
```

```python
flights = flights.assign(Year = flights['flight_date'].dt.year,
               Month = flights['flight_date'].dt.month)
```

```python
flights.columns = [col.lower() for col in flights.columns]
```

```python
flights[flights['cancelled'] | flights['diverted']].groupby('destination_state_name')\
.size().sort_values(ascending = False).plot.bar(figsize=(12,4));
```

```python
flights['distance'] = flights['distance'].str.replace(' miles', '').astype('int')
```

```python
def calc_price(miles, year, month):
    mile_price = fuel[str(year)][str(month)]
    return mile_price * miles
```

```python
flights['cost'] = flights[['distance', 'year', 'month']].apply(lambda row: calc_price(row[0], row[1], row[2]), axis = 1)
```
