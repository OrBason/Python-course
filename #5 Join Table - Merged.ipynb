# JOIN

```python
att.merge(att_new)   #join between (att and att_new) the table without changing the table - with a simple key
```

```python
heroes.merge(att_new,left_on='name',right_on='hero_names')
#connecting between two different column name as a key
#invisibility is from the att_new 
#if we want to get with 2 keys so we need to bring a list (left_on=['name','id'])
```

```python
heroes.merge(att_new,left_on='name',right_on='hero_names')[['hero_id','Gender','Race','Invisibility']]
#with select specific column
```

```python
heroes.merge(att_new,left_on='name',right_on='hero_names',how = 'outer')[['hero_id','Gender','Race','Invisibility']]
#Outer like full join in sql
#
```

```python
heroes.merge(att_new,left_on='name',right_on='hero_names',how = 'outer',indicator=True)
#Indicator=True give us a merge column and what merged
```

```python
heroes.merge(att_new,
             left_on='name',
             right_on='hero_names',
             how = 'outer',indicator=True).query('_merge !="both"')

#Indicator=True give us a merge column and what merged
```
