```python
def price_grade(p):    #defining p
    if p < 80:
        return 'Cheap'
    elif p < 120:
        return 'Medium'
    else:
        return 'Expensive'
```

```python
price_grade(114)   #Simple Use the function
```

```python
nyc['price'].apply(price_grade) 
#Use with a real case on data.
#LongWay
#apply is like a CASE in SQL - We'r creating a new field
```

```python
nyc['price'].apply(lambda pr: 'Cheap' if pr< 80 else 'Medium' if pr <120 else 'Expensive')
#ShortWay in 1 row
#Use with a real case on data.
#apply is like a CASE in SQL - We'r creating a new field
```

# Metrics with group by

```python
nyc.groupby('room_type')['price'].mean()   #mean price by room_type 
```

```python
nyc.groupby(['neighbourhood_group','room_type',])['price'].mean().sort_values(ascending=False) 
#mean price for multiple dimension with ascending
```

```python
nyc.columns
```

```python
nyc.groupby(['neighbourhood_group','room_type',],as_index=False)['price']\
.mean().sort_values('price',ascending=False) 
#mean price for multiple dimension with ascending by price   #index=False is doing a table like SQL with row
```

```python
nyc.groupby('neighbourhood_group').size().sort_values(ascending=False) 
#Count rows at some column and sort_values 
```

```python
nyc.groupby('neighbourhood_group')[['price','minimum_nights']].agg(['mean','sum'])
#agg is More flexibility, we can calulate more than one measures in the group by 
```

```python
nyc.groupby('neighbourhood_group').agg({'minimum_nights':'mean'
                                        ,'id':'count'
                                        ,'price':['median','mean']
                                        ,'number_of_reviews':'sum'})
#agg is More flexibility, we can calulate more than one measures in the group by 
```

```python
nyc.groupby('neighbourhood_group').agg(AVG_MIN_NIGHT = ('minimum_nights','mean')
                                        ,Total_Listing = ('id','count')
                                        ,median_price = ('price','median')
                                        )
#agg is More flexibility, we can calulate more than one measures in the group by
```
